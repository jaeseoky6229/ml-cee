#### **2 - Vectorized Code (Vectorization)**

Applications often need to handle large amounts of data. Using non-optimized functions (e.g., for-loops) can slow down the entire algorithm. **Vectorization** is a technique that utilizes optimized low-level routines (implemented in C/Fortran inside NumPy) to perform operations on entire arrays at once, avoiding explicit Python loops.

Key idea: **replace Python for-loops with NumPy array operations** for massive speed improvements.

##### **2.1 Dot Product — For-Loop vs Vectorized**

$$
\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^{n} a_i b_i
$$

<font color='blue'>**Example) Dot product using a for-loop (slow)**</font>

```python
import time
import numpy
import array

a = array.array('i')
for i in range(100000):
    a.append(i)
b = array.array('i')
for i in range(100000, 200000):
    b.append(i)

tic = time.process_time()
dot = 0.0
for i in range(len(a)):
    dot += a[i] * b[i]
toc = time.process_time()

print(f'For-loop dot product: {dot:.0f}')
print(f'Computation time: {(toc - tic)*1000:.4f} ms')
```

<font color='blue'>**Example) Dot product using NumPy vectorization (fast)**</font>

```python
n_tic = time.process_time()
dot_np = numpy.dot(a, b)
n_toc = time.process_time()

print(f'NumPy dot product: {dot_np}')
print(f'Computation time: {(n_toc - n_tic)*1000:.4f} ms')
```

##### **2.2 Outer Product — For-Loop vs Vectorized**

<font color='blue'>**Example) Outer product using a for-loop (slow)**</font>

```python
a = array.array('i')
for i in range(200):
    a.append(i)
b = array.array('i')
for i in range(200, 400):
    b.append(i)

tic = time.process_time()
outer = numpy.zeros((200, 200))
for i in range(len(a)):
    for j in range(len(b)):
        outer[i][j] = a[i] * b[j]
toc = time.process_time()

print(f'For-loop outer product shape: {outer.shape}')
print(f'Computation time: {(toc - tic)*1000:.4f} ms')
```

<font color='red'>**Exercise 2.1) Calculate the outer product using NumPy's vectorization (`numpy.outer`)**</font>

```python
n_tic = time.process_time()
outer_product = numpy.outer(a, b)
n_toc = time.process_time()

print(f'NumPy outer product shape: {outer_product.shape}')
print(f'Computation time: {(n_toc - n_tic)*1000:.4f} ms')
```

##### **2.3 Element-wise Product — For-Loop vs Vectorized**

<font color='blue'>**Example) Element-wise product using a for-loop (slow)**</font>

```python
a = array.array('i')
for i in range(500000):
    a.append(i)
b = array.array('i')
for i in range(500000, 1000000):
    b.append(i)

tic = time.process_time()
vector = numpy.zeros(500000)
for i in range(len(a)):
    vector[i] = a[i] * b[i]
toc = time.process_time()

print(f'For-loop element-wise product (first 5): {vector[:5]}')
print(f'Computation time: {(toc - tic)*1000:.4f} ms')
```

<font color='red'>**Exercise 2.2) Calculate the element-wise product using NumPy's vectorization (`numpy.multiply`)**</font>

```python
n_tic = time.process_time()
vector_np = numpy.multiply(a, b)
n_toc = time.process_time()

print(f'NumPy element-wise product (first 5): {vector_np[:5]}')
print(f'Computation time: {(n_toc - n_tic)*1000:.4f} ms')
```

##### **3.5 Transpose**

For matrix $A_{n \times m}$, the transpose $A^T$ has shape $m \times n$.

Properties of transpose:
1. $(A^T)^T = A$
2. $(A+B)^T = A^T + B^T$
3. $(cA)^T = cA^T$
4. $(AB)^T = B^T A^T$

<font color='blue'>**Example) Transpose a matrix**</font>

```python
A = np.array([[1, 2, 3], [4, 5, 6]])
print('A:')
print(A)
print(f'Shape: {A.shape}')

print('\nA^T:')
A_transpose = A.T
print(A_transpose)
print(f'Shape: {A_transpose.shape}')
```

<font color='red'>**Exercise 3.4) Verify the transpose property $(AB)^T = B^T A^T$**</font>

$$
A = \begin{bmatrix} 2 & 3 & 4 \\ 5 & 6 & 7 \\ 8 & 9 & 10 \end{bmatrix}, \quad
B = \begin{bmatrix} 2 & 4 & 6 \\ 8 & 10 & 12 \\ 14 & 16 & 18 \end{bmatrix}
$$

**Note**: $AB$ uses matrix multiplication (`@`), not element-wise product.

```python
A = np.arange(2, 11).reshape(3, 3)
B = np.arange(2, 20, 2).reshape(3, 3)

left_side = (A @ B).T
right_side = B.T @ A.T

print('Left side (AB)^T:')
print(left_side)
print('\nRight side B^T @ A^T:')
print(right_side)
print(f'\nAre they equal? {np.allclose(left_side, right_side)}')
```

$$

##### **4.3 Vector Products**

| Product Type | Operation | Result Shape |
|---|---|---|
| Element-wise | $a \odot c$ | Same as input |
| Dot product | $a^T c$ | Scalar |
| Outer product | $a c^T$ | $(n \times m)$ matrix |

<font color='blue'>**Example) Three types of vector products**</font>

$$
a = \begin{bmatrix} 2 \\ 1 \end{bmatrix}, \quad c = \begin{bmatrix} 1 \\ 3 \end{bmatrix}
$$

```python
a = np.array([[2], [1]])   # column vector (2,1)
c = np.array([[1], [3]])   # column vector (2,1)

# Element-wise product
elem_prod = a * c
print(f'Element-wise product:\n{elem_prod}')

# Dot product: a^T @ c → scalar
dot_prod = a.T @ c
print(f'\nDot product: {dot_prod}')

# Outer product: a @ c^T → matrix
outer_prod = a @ c.T
print(f'\nOuter product:\n{outer_prod}')
```

#### **5 - Matrix Operations & Linear Transformation**

##### **5.1 Matrix-Vector Multiplication: Row-wise**

$$
\begin{bmatrix} a_1 & a_2 \\ b_1 & b_2 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \end{bmatrix}
= \begin{bmatrix} \mathbf{row_1} \cdot \mathbf{x} \\ \mathbf{row_2} \cdot \mathbf{x} \end{bmatrix}
$$

<font color='blue'>**Example) Row-wise matrix-vector multiplication**</font>

$$
M = \begin{bmatrix} 3 & 0 \\ -2 & 1 \end{bmatrix}, \quad \vec{x} = \begin{bmatrix} -1 \\ 3 \end{bmatrix}
$$

```python
M = np.array([[3, 0], [-2, 1]])
x = np.array([[-1], [3]])

# Standard multiplication
Mx = M @ x
print(f'M @ x =\n{Mx}')

# Row-wise: each row dotted with x
row1_dot = M[0, :].dot(x)
row2_dot = M[1, :].dot(x)
Mx_row = np.vstack([row1_dot, row2_dot])
print(f'\nRow-wise result =\n{Mx_row}')
```

##### **5.2 Matrix-Vector Multiplication: Column-wise**

$$
M\vec{x} = x_1 \cdot \mathbf{col_1} + x_2 \cdot \mathbf{col_2}
$$

<font color='red'>**Exercise 5.1) Compute $M\vec{x}$ using column-wise interpretation and plot the result**</font>

$$
M = \begin{bmatrix} 3 & 0 \\ -2 & 1 \end{bmatrix}, \quad \vec{x} = \begin{bmatrix} -1 \\ 3 \end{bmatrix}
$$

```python
M = np.array([[3, 0], [-2, 1]])
x = np.array([[-1], [3]])

# Extract columns
M_col1 = M[:, 0]   # first column
M_col2 = M[:, 1]   # second column

# Column-wise multiplication: x1 * col1 + x2 * col2
M_col1x = -1 * M_col1
M_col2x = 3 * M_col2

Mx_col = M_col1x + M_col2x
print(f'Column-wise result = {Mx_col}')
print(f'Standard M@x       = {(M@x).flatten()}')
print(f'Match: {np.allclose(Mx_col, (M@x).flatten())}')
```

##### **5.3 Determinant**

For a $2 \times 2$ matrix:

$$
\det \begin{bmatrix} a & b \\ c & d \end{bmatrix} = ad - bc
$$

The determinant represents the **area scaling factor** of the linear transformation. If $\det(A) = 0$, the matrix is **singular** (not invertible). -> 역행렬이 존재한다라는 뜻

<font color='blue'>**Example) Determinant and its geometric meaning**</font>

```python
# Non-singular matrix
A = np.array([[2, 3], [1, 3]])
print(f'A =\n{A}')
print(f'det(A) = {np.linalg.det(A):.4f}')

# Singular matrix (linearly dependent columns)
B = np.array([[2, 3], [4, 6]])
print(f'\nB =\n{B}')
print(f'det(B) = {np.linalg.det(B):.4f}  → singular!')
```

##### **5.4 Inverse Matrix**

If $AB = BA = I$, then $B = A^{-1}$.

Properties:
1. $(AB)^{-1} = B^{-1}A^{-1}$
2. $(A^T)^{-1} = (A^{-1})^T$

<font color='blue'>**Example) Compute and verify an inverse matrix**</font>

```python
A = np.array([[1, 2], [3, 0]])
A_inv = np.linalg.inv(A)

print(f'A =\n{A}')
print(f'\nA_inv =\n{A_inv}')
print(f'\nA @ A_inv =\n{np.round(A @ A_inv, 10)}')
```

<font color='red'>**Exercise 5.2) Verify the property $(AB)^{-1} = B^{-1}A^{-1}$**</font>

$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 0 \end{bmatrix}, \quad
B = \begin{bmatrix} 3 & -1 \\ -3 & 2 \end{bmatrix}
$$

```python
A = np.array([[1, 2], [3, 0]])
B = np.array([[3, -1], [-3, 2]])

# Left side: (AB)^{-1}
left_side = np.linalg.inv(A @ B)

# Right side: B^{-1} A^{-1}
right_side = np.linalg.inv(B) @ np.linalg.inv(A)

print('Left side (AB)^{-1}:')
print(left_side)
print('\nRight side B^{-1} A^{-1}:')
print(right_side)
print(f'\nAre they equal? {np.allclose(left_side, right_side)}')
```

#### **6 - Eigenvalues and Eigenvectors**

##### **6.1 Geometric Intuition**

Given a matrix $A$, most vectors **change direction** when multiplied by $A$. However, some special vectors only get **scaled** (stretched or shrunk) — these are the **eigenvectors**, and the scaling factor is the **eigenvalue**.

$$
A\mathbf{v} = \lambda \mathbf{v}
$$

- $\lambda$ = **eigenvalue** (how much the vector is stretched)
- $\mathbf{v}$ = **eigenvector** (direction that doesn't change)

<font color='blue'>**Example) Visualize linear transformation and identify eigenvectors**</font>

$$
A = \begin{bmatrix} 3 & 1 \\ 0 & 2 \end{bmatrix}
$$

```python
A = np.array([[3, 1], [0, 2]])

# Test several input vectors
test_vectors = np.array([[1, 0], [1, 1], [0, 1], [-1, 1], [1, -1]])

fig, axes = plt.subplots(1, len(test_vectors), figsize=(20, 4))

for idx, x_in in enumerate(test_vectors):
    x_out = A @ x_in
    ax = axes[idx]
    ax.axhline(0, color='gray', linewidth=0.5); ax.axvline(0, color='gray', linewidth=0.5)
    ax.quiver(0, 0, x_in[0], x_in[1], angles='xy', scale_units='xy', scale=1, color='blue', label='x', linewidth=2)
    ax.quiver(0, 0, x_out[0], x_out[1], angles='xy', scale_units='xy', scale=1, color='red', label='Ax', linewidth=2)
    ax.set_xlim(-4, 5); ax.set_ylim(-4, 5); ax.set_aspect('equal')
    ax.legend(fontsize=9); ax.grid(True, alpha=0.3)
    ax.set_title(f'x={list(x_in)}', fontsize=11)

plt.suptitle('Which vectors only get scaled (same direction)?', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

##### **6.2 Computing Eigenvalues and Eigenvectors in Python**

```python
A = np.array([[3, 1], [0, 2]])

eigenvalues, eigenvectors = np.linalg.eig(A)

print(f'Eigenvalues: {eigenvalues}')
print(f'\nEigenvectors (as columns):\n{eigenvectors}')

# Verify: A @ v = lambda * v
for i in range(len(eigenvalues)):
    v = eigenvectors[:, i]
    lam = eigenvalues[i]
    print(f'\nA @ v{i+1} = {A @ v}')
    print(f'λ{i+1} * v{i+1} = {lam * v}')
    print(f'Match: {np.allclose(A @ v, lam * v)}')
```

##### **6.3 Eigenvalue Decomposition**

A square matrix $A$ with $n$ linearly independent eigenvectors can be decomposed as:

$$
A = V \Lambda V^{-1}
$$

where $V$ is the matrix of eigenvectors (columns) and $\Lambda$ is the diagonal matrix of eigenvalues.

This is useful because:
- $A^n = V \Lambda^n V^{-1}$ (efficient power computation)
- $A^{-1} = V \Lambda^{-1} V^{-1}$ (efficient inverse)

<font color='red'>**Exercise 6.1) Perform eigenvalue decomposition and verify $A = V \Lambda V^{-1}$**</font>

$$
A = \begin{bmatrix} 1 & 0 \\ 1 & 3 \end{bmatrix}
$$

```python
A = np.array([[1, 0], [1, 3]])

# Step 1: Get eigenvalues and eigenvectors
D, V = np.linalg.eig(A)

# Step 2: Create diagonal matrix of eigenvalues
Lambda = np.diag(D)

# Step 3: Compute V^{-1}
V_inv = np.linalg.inv(V)

# Step 4: Reconstruct A = V @ Lambda @ V_inv
A_reconstructed = V @ Lambda @ V_inv

print(f'Original A:\n{A}')
print(f'\nEigenvalues: {D}')
print(f'\nReconstructed A:\n{np.round(A_reconstructed, 10)}')
print(f'\nMatch: {np.allclose(A, A_reconstructed)}')
```

##### **6.4 Application: Efficient Matrix Power using Eigendecomposition**

<font color='blue'>**Example) Compare computation time for $A^{50}$: direct vs eigendecomposition**</font>

```python
import time

A = np.array([[1, 0], [1, 0.5]])

# Method 1: Direct repeated multiplication
tic = time.perf_counter()
result1 = np.eye(2)
for _ in range(50):
    result1 = result1 @ A
toc = time.perf_counter()
print(f'Direct method: {(toc-tic)*1e6:.1f} μs')
print(f'Result:\n{result1}\n')

# Method 2: Eigendecomposition A^n = V @ Lambda^n @ V_inv
tic = time.perf_counter()
D, V = np.linalg.eig(A)
Lambda_50 = np.diag(D**50)
V_inv = np.linalg.inv(V)
result2 = V @ Lambda_50 @ V_inv
toc = time.perf_counter()
print(f'Eigendecomp method: {(toc-tic)*1e6:.1f} μs')
print(f'Result:\n{np.real(result2)}')
```

> 지금은 행렬 자체가 작아서 direct method가 더 빠른데, 행렬이 커지고 반복 횟수가 더 많아지면 eigencomp가 더 빨라진다.

A = V Λ V⁻¹<br>
<br>
A² = A @ A<br>
   = (V Λ V⁻¹) @ (V Λ V⁻¹)<br>
   = V Λ (V⁻¹ @ V) Λ V⁻¹<br>
   = V Λ  I  Λ V⁻¹<br>
   = V Λ² V⁻¹<br>
<br>
A³ = A² @ A<br>
   = (V Λ² V⁻¹) @ (V Λ V⁻¹)<br>
   = V Λ² (V⁻¹ @ V) Λ V⁻¹<br>
   = V Λ²  I  Λ V⁻¹<br>
   = V Λ³ V⁻¹<br>

<font color='red'>**Exercise 6.2) Compute the inverse of $A$ using eigendecomposition: $A^{-1} = V \Lambda^{-1} V^{-1}$**</font>

$$
A = \begin{bmatrix} 1 & 0 \\ 1 & 0.5 \end{bmatrix}
$$

Compare with `np.linalg.inv(A)` to verify.

```python
A = np.array([[1, 0], [1, 0.5]])

# Method 1: Direct inverse
A_inv_direct = np.linalg.inv(A)
print(f'Direct inverse:\n{A_inv_direct}\n')

# Method 2: Eigendecomposition
D, V = np.linalg.eig(A)
D_inv = 1 / D
Lambda_inv = np.diag(D_inv)
V_inv = np.linalg.inv(V)
A_inv_eig = V @ Lambda_inv @ V_inv

print(f'Eigendecomp inverse:\n{np.real(A_inv_eig)}\n')
print(f'Match: {np.allclose(A_inv_direct, A_inv_eig)}')
```
